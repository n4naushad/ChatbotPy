## 1. Data Import and Cleaning

* Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set visualization style
sns.set(style="whitegrid")
```

* Load the dataset into a pandas DataFrame

```python
# Load the dataset into a pandas DataFrame
lv_file_path = 'G:\\M Tech - Study\\Class\\01 - SEM 1\\Mini Project\\PDS Project\\Raw_Medical_dataset.csv'

# Assuming the dataset file is already loaded
df = pd.read_csv(lv_file_path)
```

* Perform initial data inspection

```python
# Check the 'shape' of the data
print()
print("Shape of the dataset:", df.shape)
print()

# Check 'data types' of each column
print("Data types of each column:\n__________________________" )
print(df.dtypes)
print()

# 'Summary statistics' for numerical columns
print("Summary statistics:\n___________________")
print(df.describe())
print()
```

* Identify and Handling missing values

```python
# Check for missing values
print("Missing values in the columns:\n______________________________")
print(df.isnull().sum())

# Handle missing values
df_cleaned = df.dropna()   # dropna() function is used to remove missing values or dropping rows from a DataFrame
```

* Detect and remove duplicate rows

```python
# Check for duplicates
print("Number of duplicate rows before removing Duplicates:", df_cleaned.duplicated().sum())

# Remove duplicates
df_cleaned = df_cleaned.drop_duplicates()

# Check for duplicates
print("Number of duplicate rows after  removing Duplicates:", df_cleaned.duplicated().sum())
```

## 2. Exploratory Data Analysis (EDA)

## •	Descriptive Statistics

* Summary statistics for numerical columns

```python
print("Numerical summary:\n__________________")
print(df_cleaned.describe())
print()
```

* Summary statistics for categorical columns

```python
print("Categorical summary:\n____________________")
print(df_cleaned.describe(include=['object']))
print()
```

## •	Data Visualization

* Histograms for numerical features

```python
# Create histograms or density plots for numerical features

#df_cleaned.hist(bins=30, figsize=(10, 8))
df_cleaned.hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

* Bar plots for categorical features

```python
# Create bar plots for categorical features

for column in df_cleaned.select_dtypes(include=['object']).columns:
    sns.countplot(data=df_cleaned, x=column)
    plt.title(f'Bar Plot of {column}')
    plt.show()
```

* Box plots to identify outliers

```python
# Create scatter plots to explore relationships between numerical features

for column in df_cleaned.select_dtypes(include=[np.number]).columns:
    sns.boxplot(data=df_cleaned, y=column)
    plt.title(f'Box Plot of {column}')
    plt.show()
```

* Scatter plots to explore relationships

```python
sns.pairplot(df_cleaned)
plt.suptitle('Pairplot of Features')
plt.show()
```

* Heatmap for correlations

```python
# Assuming df_cleaned is your DataFrame
# Attempt to convert columns to numeric, coercing errors to NaN
df_numeric = df_cleaned.apply(pd.to_numeric, errors='coerce')

# Drop non-numeric columns (if any) or columns that could not be converted
df_numeric = df_numeric.dropna(axis=1, how='all')

# Now calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## •	Group Analysis

* Group-by operations based on categorical features

```python
# Assuming df_cleaned is your cleaned DataFrame

# Step 1: Select only the numeric columns
numeric_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_numeric = df_cleaned[numeric_columns]

# Step 2: Group by 'Job' and calculate the mean for numeric columns
grouped_data = df_numeric.groupby(df_cleaned['Job']).mean()

# Step 3: Display the grouped data
print("\nGrouped Data Summary:\n", grouped_data)
```

# 3. Advanced Python Techniques

## •	Lambda Functions

* Use lambda functions for simple data transformations

```python
# Example: Create a new column categorizing numerical data into bins
# Vitamin D levels
# Deficiency: Less than 20 ng/mL
# Insufficiency: 20 to 29 ng/mL
# Sufficiency: 30 to 100 ng/mL
# Potential toxicity: Greater than 100 ng/mL

df_cleaned['Deficiency_category'] = df_cleaned['VitD_levels'].apply(lambda x: 'Deficiency' if x < 20 else 'Insufficiency' if x < 30 else 'Sufficiency')
print(df_cleaned['Deficiency_category'] )
```

```python
# Example: Create a new column categorizing numerical data into bins
df_cleaned['size_category'] = df_cleaned['Age'].apply(lambda x: 'Child' if x < 19 else 'Adult' if x < 65 else 'Elders')
print(df_cleaned['size_category'] )
```

## •	User-Defined Functions

```python
# Example: Function to categorize age
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

df_cleaned['age_category'] = df_cleaned['Age'].apply(categorize_age)
print(df_cleaned)
```
